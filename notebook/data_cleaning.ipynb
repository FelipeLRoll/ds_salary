```python
# Create a dictionary mapping degree
degree_mapping = {
    'M': 'Master',
    'P': 'Postgraduate',
    'na': 'None'
}
```

```python
# Create a dictionary mapping size ranges 
size_mapping = {
    '1 - 50': 25,
    '51 - 200': 125,
    '201 - 500': 350,
    '501 - 1000': 750,
    '1001 - 5000': 3000,
    '5001 - 10000': 7500,
    '10000+': 10000
}
```

```python
# Create a dictionary mapping revenue
revenue_mapping = {
    'Less than $1 million (USD)': 1000000,
    '$1 to $5 million (USD)': 3000000,
    '$5 to $10 million (USD)': 7500000,
    '$10 to $25 million (USD)': 17500000,
    '$25 to $50 million (USD)': 37500000,
    '$50 to $100 million (USD)': 75000000,
    '$100 to $500 million (USD)': 300000000,
    '$500 million to $1 billion (USD)': 500000000,
    '$1 to $2 billion (USD)': 1500000000,
    '$2 to $5 billion (USD)': 3500000000,
    '$5 to $10 billion (USD)': 7500000000,
    '$10+ billion (USD)': 10000000000,
    'Unknown / Non-Applicable': 0 
}
```

```python
# Create a dictionary mapping skils needed columns
skills_mapping = {
    0: 'No',
    1: 'Yes'
}
```

```python
#renaming columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# Rename the 'company_txt' column to 'company_name'
df = df.rename(columns={'company_txt': 'company'})
```

```python
# Create a new column 'location_state' with the state part of 'location'
df['job_location'] = df['location'].str.split(',').str[-1].str.strip()

# Replace abbreviations in 'job_location' 
df['job_location'] = df['job_location'].replace(abbrevs)
```

```python
# Create a new column 'location_city' with the city part of 'location'
df['job_location_city'] = df['location'].str.split(',').str[0].str.strip()
```

```python
# Create a new column 'headquarters_state' with the state part of 'headquarters'
df['company_headquarters'] = df['headquarters'].str.split(',').str[-1].str.strip()

# Replace abbreviations in 'company_headquarters' 
df['company_headquarters'] = df['company_headquarters'].replace(abbrevs)
```

```python
# Create a new column 'lheadquarters_city' with the city part of 'headquarters'
df['company_headquarters_city'] = df['headquarters'].str.split(',').str[0].str.strip()
```

```python
# Create a new column 'same_place' to check if location_city and headquarters_city are the same
df['same_place'] = (df['job_location_city'].str.strip().str.lower() == df['company_headquarters_city'].str.strip().str.lower()) & (df['job_location'].str.strip().str.lower() == df['company_headquarters'].str.strip().str.lower())
```

```python
#converts the avg_salary(k) to yearly if it is hourly salary
df['avg_salary(k)'] = np.where(df['hourly'] == 1, df['avg_salary(k)'] * 2.08, df['avg_salary(k)'])
```

```python
#create column to show avg salary per hour
#2080 represents the average number of working hours in a year considering a standard 40-hour workweek and 52 weeks in a year. Dividing the annual salary by 2.08 gives an approximate hourly rate.
df['avg_salary_hour($)'] = round(df['avg_salary(k)'] / 2.08, 2)
```

```python
#create column to show avg salary per month
df['avg_salary_month($)'] = round(df['avg_salary(k)'] / 12)  * 1000
```

```python
#create column to show avg salary per year 
df['avg_salary_year($)'] = round(df['avg_salary(k)'] * 1000 ) 
```

```python
# Remove the trailing space in the 'size' column
df['size'] = df['size'].str.rstrip()

#converting company size(number of workers) to labels = 1-50= micro,  51-200 = small, 201-500 small-medium, 501-1000 = medium,  1001-5000 medium-big,  5001-10000 = big, 10000+ = large
df['number_of_workers'] = df['size'].replace(size_mapping)
```

```python
#revenue mapping
df['revenue(USD)'] = df['revenue'].replace(revenue_mapping)
```

```python
#degree mapping
df['higher_degree'] = df['degree'].replace(degree_mapping)
```

```python
#skills mapping
df['python'] = df['python'].replace(skills_mapping)
df['spark'] = df['spark'].replace(skills_mapping)
df['aws'] = df['aws'].replace(skills_mapping)
df['excel'] = df['excel'].replace(skills_mapping)
df['sql'] = df['sql'].replace(skills_mapping)
df['sas'] = df['sas'].replace(skills_mapping)
df['keras'] = df['keras'].replace(skills_mapping)
df['scikit'] = df['scikit'].replace(skills_mapping)
df['tensor'] = df['tensor'].replace(skills_mapping)
df['hadoop'] = df['hadoop'].replace(skills_mapping)
df['tableau'] = df['tableau'].replace(skills_mapping)
df['bi'] = df['bi'].replace(skills_mapping)
df['flink'] = df['flink'].replace(skills_mapping)
df['mongo'] = df['mongo'].replace(skills_mapping)
df['google_an'] = df['google_an'].replace(skills_mapping)
df['pytorch'] = df['pytorch'].replace(skills_mapping)


#also doing the same for employer_provided
df['employer_provided'] = df['employer_provided'].replace(skills_mapping)
```

```python
# Change 'job_title_sim' column to have uppercase first letter of each word
df['job_title_sim'] = df['job_title_sim'].apply(lambda x: x.title())
```

```python
# Replace '\n' with ' ' in the 'job_description' column
df['job_description'] = df['job_description'].str.replace('\n', ' ')
```

```python
# Replace -1 with 'No Data' in string columns and 0 in numeric columns
df = df.apply(lambda x: x.replace('-1', 'No Data') if x.dtype == 'object' else x.replace(-1, 0))
```

```python
# Convert 'founded' column to string and replace 0 with 'No Data'
df['founded'] = df['founded'].astype(str).replace('0', 'No Data')
```

```python
# Create a new column 'number_competitors' with the count of competitors
df['number_competitors'] = df['competitors'].apply(lambda x: 0 if x == 'No Data' else len(x.split(',')))
```

```python

# Drop rows where 'job_title_sim' is 'na'
df = df[df['job_title_sim'].str.lower() != 'na']

# Drop rows where 'size' is 'unknown'
df = df[df['size'].str.lower() != 'unknown']

# Drop the rows where 'rating' is negative
df = df[df['rating'] >= 0]

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)
```

```python
#remove columns
df = df.drop(columns=['index', 'seniority_by_title', 'hourly', 'company_name', 'avg_salary(k)','location','headquarters','salary_estimate','size','revenue','lower_salary', 'upper_salary','degree'])
```

```python
# Define the new order of columns
new_column_order = ['job_title', 'job_title_sim', 'job_description', 'python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit', 'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an', 'higher_degree','avg_salary_hour($)', 'avg_salary_month($)', 'avg_salary_year($)', 'employer_provided','job_location','job_location_city',  'company', 'company_headquarters', 'company_headquarters_city', 'same_place', 'rating', 'number_of_workers', 'revenue(USD)', 'founded', 'type_of_ownership', 'industry', 'sector', 'age', 'competitors', 'number_competitors' ]

#Reorder the columns in the DataFrame
df = df[new_column_order]
```
